```python
user_input = input()
pixel_values = user_input.split()

pixel_data = [tuple(map(int, pixel.split(','))) for pixel in pixel_values]

# Convert RGB pixels to grayscale
grayscale_values = [int(0.2989 * pixel[0] + 0.5870 * pixel[1] + 0.1140 * pixel[2]) for pixel in pixel_data]

# Calculate the average grayscale value
avg_grayscale = sum(grayscale_values) / len(grayscale_values)

threshold = 100

if avg_grayscale > threshold:
    print("day")
else:
    print("night")
```
